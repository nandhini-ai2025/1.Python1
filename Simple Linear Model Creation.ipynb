```python
#import pandas library which contains all the function related to database
import pandas as pd
```

```python
#read the data file and create the dataset
dataset=pd.read_csv("Salary_Data.csv")
```

```python
dataset
```

```python
#create the input(independent) and output(dependent) variables
independent=dataset[["YearsExperience"]]
```

```python
independent
```

```python
dependent=dataset[["Salary"]]
dependent
```

```python
#create the train dataset and create the test dataset
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#select the linear regression model from sklearn library
from sklearn.linear_model import LinearRegression
regressor= LinearRegression()
#train the model using train dataset
regressor.fit(X_train,y_train)
```

```python
#calculate the weight and bais value
weight=regressor.coef_
weight
```

```python
bais=regressor.intercept_
bais
```

```python
#now we can predict the model
y_pred=regressor.predict(X_test)
```

```python
#do the evaluation the model
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
#our model predicts nearly 97%,which is show it's good model,now save the model
import pickle
#file name of the linear regression model
filename="finalized_model_linear.sav"
```

```python
#save the model
pickle.dump(regressor,open(filename,'wb'))
```

```python
#loaded the saved model into the object
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))
#from the loaded model predicts the salary based on the year of experience
result=loaded_model.predict([[14]])
```
